## Import the required libraries

```
import os
import glob
import random
import shutil
from pathlib import Path
```

## Prepare the training data

```
ORIG_IMAGES_DIR = os.path.join('/data/catflap/catflap_pictures_orig/yolox/yolox_ds/')
ORIG_ANNOTATIONS_DIR = os.path.join('/data/catflap/catflap_pictures_orig/yolox/yolox_annotations/')

VOC_IMAGES_DIR = os.path.join('/data/catflap/catflap_pictures/yolox/voc_ds/')
VOC_ANNOTATIONS_DIR = os.path.join('/data/catflap/catflap_pictures/yolox/voc_annotations/')
VOC_LABEL_DIR = os.path.join('/data/catflap/catflap_pictures/yolox/voc_main/')

list_ds = glob.glob(ORIG_IMAGES_DIR+"/*.png")
random.shuffle(list_ds)

image_count = len(list_ds)
train_size = int(0.8 * image_count)
val_size = int(0.2 * image_count)

train_ds = list_ds[:train_size]
val_ds = list_ds[train_size:]

print(image_count)
print(len(train_ds))
print(len(val_ds))
```

```
if not os.path.exists(VOC_IMAGES_DIR):
    os.mkdir(VOC_IMAGES_DIR)
if not os.path.exists(VOC_ANNOTATIONS_DIR):
    os.mkdir(VOC_ANNOTATIONS_DIR)
if not os.path.exists(VOC_LABEL_DIR):
    os.mkdir(VOC_LABEL_DIR)
if os.path.exists(VOC_LABEL_DIR+"train.txt"):
    os.remove(VOC_LABEL_DIR+"train.txt")
if os.path.exists(VOC_LABEL_DIR+"val.txt"):
    os.remove(VOC_LABEL_DIR+"val.txt")
```

```
i = 1

f = open(VOC_LABEL_DIR+"val.txt", "w")
for filepath in val_ds:
  print(i,':', filepath)
  p = Path(filepath)
  #print(p.parts[-1])
  #print(p.stem)
  filepath_annotation = ORIG_ANNOTATIONS_DIR + p.stem + ".xml"
  shutil.copy2(filepath, VOC_IMAGES_DIR)
  shutil.copy2(filepath_annotation, VOC_ANNOTATIONS_DIR)
  f.write(p.stem)
  f.write('\n')
  i = i + 1
f.close()
```

```
i = 1

f = open(VOC_LABEL_DIR+"train.txt", "w")
for filepath in train_ds:
  print(i,':', filepath)
  p = Path(filepath)
  #print(p.parts[-1])
  #print(p.stem)
  filepath_annotation = ORIG_ANNOTATIONS_DIR + p.stem + ".xml"
  shutil.copy2(filepath, VOC_IMAGES_DIR)
  shutil.copy2(filepath_annotation, VOC_ANNOTATIONS_DIR)
  f.write(p.stem)
  f.write('\n')
  i = i + 1
f.close()
```
